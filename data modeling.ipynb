```python
import pandas as pd
import pyodbc

# Replace the connection string with your actual SQL Server connection string
conn_str = 'DRIVER={SQL Server};SERVER=LAPTOP-REU1GLF5\SQLEXPRESS;DATABASE=Github_base'

# Establish a connection to the database
conn = pyodbc.connect(conn_str)
cursor = conn.cursor()
```

```python
# Drop the "Repositories" table if it exists
cursor.execute("IF OBJECT_ID('Repository_Languages', 'U') IS NOT NULL DROP TABLE Repository_Languages")
cursor.execute("IF OBJECT_ID('Repository_Contributors', 'U') IS NOT NULL DROP TABLE Repository_Contributors")
cursor.execute("IF OBJECT_ID('Users', 'U') IS NOT NULL DROP TABLE Users")
cursor.execute("IF OBJECT_ID('Repositories', 'U') IS NOT NULL DROP TABLE Repositories")
cursor.execute("IF OBJECT_ID('Languages', 'U') IS NOT NULL DROP TABLE Languages")
cursor.execute("IF OBJECT_ID('Contributors', 'U') IS NOT NULL DROP TABLE Contributors")
conn.commit()
```

```python
# Create the table if it doesn't exist
cursor.execute('''
CREATE TABLE Repositories (
    private BIT,
    created_at DATE,
    updated_at DATE,
    size INT,
    stars INT,
    open_issues_count INT,
    forks_count INT,
    owner_type VARCHAR(50),
    repo_id INT PRIMARY KEY
)
''')
conn.commit()

# Insert data into the table
for row in df_repo.itertuples(index=False):
    cursor.execute("INSERT INTO Repositories (private, created_at, updated_at, size, stars, open_issues_count, forks_count, owner_type, repo_id) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?)", 
                   row.private, row.created_at, row.updated_at, row.size, row.stargazers_count, row.open_issues_count, row.forks_count, row.owner_type, row.repo_id)
    conn.commit()
```

```python
# Create the Languages table if it doesn't exist
cursor.execute('''
CREATE TABLE Languages (
    language_id INT PRIMARY KEY,
    language_name VARCHAR(50),
    frequence INT
)
''')
conn.commit()
```

```python
unique_languages = table_langue['languages'].unique()
for idx, language in enumerate(unique_languages):
    cursor.execute("INSERT INTO Languages (language_id, language_name, frequence) VALUES (?, ?, ?)", idx + 1, language, int(table_langue[table_langue['languages'] == language]['language_byte'].sum()))
    conn.commit()
```

```python
# Create the Repository_Languages table with foreign keys to Repositories and Languages tables
cursor.execute('''
CREATE TABLE Repository_Languages (
    repo_id INT,
    language_id INT,
    PRIMARY KEY (repo_id, language_id),
    FOREIGN KEY (repo_id) REFERENCES Repositories(repo_id),
    FOREIGN KEY (language_id) REFERENCES Languages(language_id)
)
''')
conn.commit()
```

```python
# Map the languages to their respective language_id and insert into the Repository_Languages table
languages_mapping = {language: idx + 1 for idx, language in enumerate(unique_languages)}
repo_lang_data = []

for row in table_langue.itertuples(index=False):
    language, repo_id, _ = row
    language_id = languages_mapping[language]
    repo_lang_data.append((repo_id, language_id))

for data in repo_lang_data:
    cursor.execute("INSERT INTO Repository_Languages (repo_id, language_id) VALUES (?, ?)", data[0], data[1])
conn.commit()
```

```python
cursor.execute('''
CREATE TABLE Contributors (
    contributor_id INT PRIMARY KEY,
    Contributeur VARCHAR(50),
    nombre_de_contribution INT
)
''')
conn.commit()
```

```python
# Insert data into the Contributors table (only unique contributor names)
unique_contributors = df2['Contributors'].unique()

for idx, contributor in enumerate(unique_contributors):
    cursor.execute("INSERT INTO Contributors (contributor_id, Contributeur, nombre_de_contribution) VALUES (?, ?, ?)", 
                   idx + 1, contributor, int(df2[df2['Contributors'] == contributor]['Contributors_frequency'].sum()))
conn.commit()
```

```python
# Create the Repository_Contributors table with foreign keys to Repositories and Contributors tables
cursor.execute('''
CREATE TABLE Repository_Contributors (
    repo_id INT,
    contributor_id INT,
    PRIMARY KEY (repo_id, contributor_id),
    FOREIGN KEY (repo_id) REFERENCES Repositories(repo_id),
    FOREIGN KEY (contributor_id) REFERENCES Contributors(contributor_id)
)
''')
conn.commit()
```

```python
# Map the contributors to their respective contributor_id and insert into the Repository_Contributors table
contributors_mapping = {contributor: idx + 1 for idx, contributor in enumerate(unique_contributors)}
repo_contrib_data = []

for row in df2.itertuples(index=False):
    repo_id = row.repo
    contributor = row.Contributors
    contributor_id = contributors_mapping[contributor]
    repo_contrib_data.append((repo_id, contributor_id))
    
for data in repo_contrib_data:
    cursor.execute("INSERT INTO Repository_Contributors (repo_id, contributor_id) VALUES (?, ?)", data[0], data[1])
    conn.commit()
```

```python
cursor.execute('''
CREATE TABLE Users (
    user_id INT PRIMARY KEY,
    user_name VARCHAR(50),
    repo_id INT,
    FOREIGN KEY (repo_id) REFERENCES Repositories(repo_id)
)
''')
conn.commit()
```

```python
# Insert data into the "Users" table
unique_users = table_user['full_name'].unique()

for idx, user in enumerate(unique_users):
    cursor.execute("INSERT INTO Users (user_id, user_name, repo_id) VALUES (?, ?, ?)", idx + 1, user, 
                   int(table_user[table_user['full_name'] == user]['repo_id'].values[0]))
conn.commit()
```
